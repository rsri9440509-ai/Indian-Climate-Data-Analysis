```python
import pandas as pd

# Load the provided dataset
try:
    df = pd.read_csv('/content/Indian_Climate_Dataset_2024_2025.csv')
    print("Dataset 'Indian_Climate_Dataset_2024_2025.csv' loaded successfully. Here are the first 5 rows:")
    display(df.head())
except FileNotFoundError:
    print("Error: Dataset not found. Please ensure the file is in the correct path.")
except Exception as e:
    print(f"An error occurred while loading the dataset: {e}")
```

```python
print("\n--- Dataset Information ---")
df.info()

print("\n--- Missing Values Count ---")
print(df.isnull().sum())

print("\n--- Descriptive Statistics ---")
display(df.describe())
```

```python
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Clean column names by removing special characters and replacing spaces with underscores
df.columns = df.columns.str.replace(r'[°%()\/]', '', regex=True).str.replace(' ', '_').str.strip()

# Display the cleaned column names and info to confirm changes
print("\n--- Cleaned Column Names ---")
print(df.columns)
print("\n--- Dataset Information After Cleaning ---")
df.info()

# Display the first few rows with cleaned data to verify
print("\n--- First 5 rows after cleaning ---")
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Group by Date to get daily averages for trend analysis
daily_avg_df = df.groupby('Date').agg({
    'Temperature_Avg_C': 'mean',
    'Humidity_': 'mean',
    'Rainfall_mm': 'sum', # Sum rainfall for daily total
    'AQI': 'mean',
    'Wind_Speed_kmh': 'mean',
    'Pressure_hPa': 'mean',
    'Cloud_Cover_': 'mean'
}).reset_index()

# Plotting trends for key meteorological parameters

plt.figure(figsize=(18, 12))

# Temperature Trend
plt.subplot(3, 2, 1)
sns.lineplot(x='Date', y='Temperature_Avg_C', data=daily_avg_df)
plt.title('Daily Average Temperature Trend (°C)')
plt.xlabel('Date')
plt.ylabel('Average Temperature (°C)')

# Humidity Trend
plt.subplot(3, 2, 2)
sns.lineplot(x='Date', y='Humidity_', data=daily_avg_df, color='green')
plt.title('Daily Average Humidity Trend (%)')
plt.xlabel('Date')
plt.ylabel('Average Humidity (%)')

# Rainfall Trend
plt.subplot(3, 2, 3)
sns.lineplot(x='Date', y='Rainfall_mm', data=daily_avg_df, color='blue')
plt.title('Daily Total Rainfall Trend (mm)')
plt.xlabel('Date')
plt.ylabel('Total Rainfall (mm)')

# AQI Trend
plt.subplot(3, 2, 4)
sns.lineplot(x='Date', y='AQI', data=daily_avg_df, color='red')
plt.title('Daily Average AQI Trend')
plt.xlabel('Date')
plt.ylabel('Average AQI')

# Wind Speed Trend
plt.subplot(3, 2, 5)
sns.lineplot(x='Date', y='Wind_Speed_kmh', data=daily_avg_df, color='purple')
plt.title('Daily Average Wind Speed Trend (km/h)')
plt.xlabel('Date')
plt.ylabel('Average Wind Speed (km/h)')

# Cloud Cover Trend
plt.subplot(3, 2, 6)
sns.lineplot(x='Date', y='Cloud_Cover_', data=daily_avg_df, color='gray')
plt.title('Daily Average Cloud Cover Trend (%)')
plt.xlabel('Date')
plt.ylabel('Average Cloud Cover (%)')

plt.tight_layout()
plt.show()
```

```python
# Group by City and calculate the mean of relevant meteorological parameters
city_avg_df = df.groupby('City').agg({
    'Temperature_Avg_C': 'mean',
    'Humidity_': 'mean',
    'Rainfall_mm': 'sum', # Sum rainfall for total comparison
    'AQI': 'mean',
    'Wind_Speed_kmh': 'mean',
    'Pressure_hPa': 'mean',
    'Cloud_Cover_': 'mean'
}).reset_index()

print("\n--- Average Meteorological Parameters by City ---")
display(city_avg_df)

# Visualize the comparison across cities for key parameters
plt.figure(figsize=(18, 12))

# Average Temperature by City
plt.subplot(2, 3, 1)
sns.barplot(x='City', y='Temperature_Avg_C', data=city_avg_df, palette='viridis', hue='City', legend=False)
plt.title('Average Temperature by City (°C)')
plt.xlabel('City')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=45, ha='right')

# Average Humidity by City
plt.subplot(2, 3, 2)
sns.barplot(x='City', y='Humidity_', data=city_avg_df, palette='magma', hue='City', legend=False)
plt.title('Average Humidity by City (%)')
plt.xlabel('City')
plt.ylabel('Average Humidity (%)')
plt.xticks(rotation=45, ha='right')

# Total Rainfall by City
plt.subplot(2, 3, 3)
sns.barplot(x='City', y='Rainfall_mm', data=city_avg_df, palette='cividis', hue='City', legend=False)
plt.title('Total Rainfall by City (mm)')
plt.xlabel('City')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=45, ha='right')

# Average AQI by City
plt.subplot(2, 3, 4)
sns.barplot(x='City', y='AQI', data=city_avg_df, palette='inferno', hue='City', legend=False)
plt.title('Average AQI by City')
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.xticks(rotation=45, ha='right')

# Average Wind Speed by City
plt.subplot(2, 3, 5)
sns.barplot(x='City', y='Wind_Speed_kmh', data=city_avg_df, palette='plasma', hue='City', legend=False)
plt.title('Average Wind Speed by City (km/h)')
plt.xlabel('City')
plt.ylabel('Average Wind Speed (km/h)')
plt.xticks(rotation=45, ha='right')

# Average Cloud Cover by City
plt.subplot(2, 3, 6)
sns.barplot(x='City', y='Cloud_Cover_', data=city_avg_df, palette='GnBu_r', hue='City', legend=False)
plt.title('Average Cloud Cover by City (%)')
plt.xlabel('City')
plt.ylabel('Average Cloud Cover (%)')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
print("\n--- Correlation Matrix of Meteorological Parameters ---")

# Select only numerical columns for correlation analysis
# Exclude 'Date', 'City', 'State', and 'AQI_Category' as they are not numerical
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Meteorological Parameters')
plt.show()
```

```python
# Visualize the relationship between AQI and Wind Speed
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wind_Speed_kmh', y='AQI', data=df, alpha=0.5, color='teal')
plt.title('Relationship between AQI and Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('AQI')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping AQI Categories by City
aqi_distribution = df.groupby(['City', 'AQI_Category']).size().unstack(fill_value=0)

print("\n--- AQI Category Distribution by City ---")
display(aqi_distribution)

# Visualizing the distribution using a stacked bar chart
aqi_distribution.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='viridis')

plt.title('AQI Category Distribution by City')
plt.xlabel('City')
plt.ylabel('Number of Days')
plt.xticks(rotation=45, ha='right')
plt.legend(title='AQI Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Actionable Insights and Conclusion

Based on the end-to-end analysis of the Indian Climate Dataset (2024-2025), we have gathered several key insights:

### Data Cleaning Summary
The dataset was remarkably clean, with no missing values. The 'Date' column was successfully converted to datetime objects, and column names were standardized for easier manipulation. This ensured a robust foundation for subsequent analysis.

### Trend Analysis Insights
*   **Temperature**: The dataset shows noticeable fluctuations in average daily temperatures, indicating seasonal variations typical for the Indian subcontinent. Further analysis could delve into specific seasonal trends.
*   **Humidity**: Daily average humidity also shows variations, often inversely correlated with temperature. High humidity periods might correspond to monsoons or specific weather events.
*   **Rainfall**: Total daily rainfall exhibits spikes, suggesting episodic heavy rainfall events rather than continuous precipitation. These spikes are crucial for understanding flood risks and water management.
*   **AQI (Air Quality Index)**: AQI trends indicate periods of varying air quality, with certain days or seasons potentially experiencing higher pollution levels. This highlights the need for air quality monitoring and mitigation strategies.
*   **Wind Speed and Cloud Cover**: These parameters show day-to-day variability, influencing other weather phenomena like temperature, humidity, and rainfall.

### City Comparison Insights
*   **Temperature Variations**: While average temperatures across cities are relatively close, slight differences can be significant for microclimates and urban planning. Some cities might consistently experience higher or lower temperatures, affecting energy consumption and heat stress.
*   **Rainfall Distribution**: Total rainfall varies significantly across cities, with some regions receiving substantially more precipitation. This has direct implications for agriculture, water resource management, and flood preparedness in different urban areas.
*   **AQI Differences**: AQI levels show variations among cities, pointing to different levels of air pollution. Cities with consistently high AQI might require targeted interventions to improve air quality and public health.
*   **Humidity, Wind Speed, and Cloud Cover**: These also show city-specific characteristics, contributing to the unique climatic profiles of each city.

### Correlation Analysis Insights
*   **Strong Positive Correlation**: As expected, `Temperature_Max_C`, `Temperature_Min_C`, and `Temperature_Avg_C` are highly positively correlated with each other, confirming consistency in temperature measurements.
*   **Temperature and Humidity**: Often, there's a negative correlation between temperature and humidity. Higher temperatures might lead to lower relative humidity, or vice-versa, depending on local conditions and time of year.
*   **Rainfall and Cloud Cover**: A positive correlation between `Rainfall_mm` and `Cloud_Cover_` is intuitive, as increased cloud cover often precedes or accompanies rainfall.
*   **AQI and Other Factors**: The correlation of AQI with other meteorological parameters (e.g., wind speed, temperature) can indicate how weather conditions influence air pollution dispersion or concentration. For instance, low wind speeds might correlate with higher AQI due to reduced pollutant dispersion.

### Actionable Recommendations
1.  **Urban Planning & Infrastructure**: Cities with higher average temperatures and lower rainfall should focus on water conservation, green infrastructure (e.g., urban forests, cool roofs), and heat-resilient building designs.
2.  **Public Health Warnings**: Implement and improve early warning systems for periods of high AQI and extreme temperatures (heatwaves). Public awareness campaigns on preventive measures are crucial.
3.  **Agricultural Planning**: Utilize rainfall and temperature trends to inform agricultural practices and crop selection in different regions, optimizing yields and minimizing climate-related losses.
4.  **Pollution Control**: For cities with consistently high AQI, investigate specific sources of pollution (industrial, vehicular, seasonal burning) and implement targeted control measures, potentially leveraging wind speed data for effective dispersion strategies.
5.  **Climate Change Adaptation**: The observed trends and city comparisons can serve as a baseline for developing region-specific climate change adaptation strategies, focusing on resilience against extreme weather events.

### Conclusion
The analysis provides a comprehensive overview of the climate patterns across major Indian cities in 2024-2025. By understanding the trends, inter-city variations, and correlations between meteorological parameters, stakeholders can make informed decisions in urban development, environmental policy, public health, and disaster preparedness to build more climate-resilient communities. Further in-depth analysis for specific cities or seasons could yield even more granular insights.
